```python
from numpy import *
import bqplot as bq
from ipywidgets import interact, interactive, Layout, Box, HBox, VBox, Label
import ipywidgets as widgets
from IPython.display import display
```

```python
class curve:
    
    t = arange(0, 2*pi, 0.001) # Time points
    a, b, c, d = 4, 8, -2.17, 6.5 
    
    # Initial setup
    def __init__(self):
        self.x = cos(self.t) + cos(self.a*self.t)/self.c + sin(self.b*self.t)/self.d 
        self.y = sin(self.t) + sin(self.a*self.t)/self.c + cos(self.b*self.t)/self.d

        # Generate a line (but does not plot it)
        self.l = bq.Lines(x=self.x, y=self.y, colors=['red'], opacities=[0.1], fill='bottom', fill_colors=['red'], fill_opacities=[0.2], \
                          scales={'x': bq.LinearScale(min=-3, max=3), 'y': bq.LinearScale(min=-2.5, max=2.5)})

        # Draw the above defined line in figure
        self.fig = bq.Figure(marks=[self.l], min_aspect_ratio = 1, max_aspect_ratio = 1)

    # Draw the equation
    def pattern(self, a=3, b=2, c=2, d=9, color='red', alpha=0.2):
        self.a, self.b, self.c, self.d = a, b, c, d
        
        # Equation for the pattern
        x = cos(self.t) + cos(self.a*self.t)/self.c + sin(self.b*self.t)/self.d 
        y = sin(self.t) + sin(self.a*self.t)/self.c + cos(self.b*self.t)/self.d

        self.l.x = x # Set x values of line
        self.l.y = y # Set y values of line
        self.l.colors = self.l.fill_colors = [color] # set colour
        self.l.fill_opacities=[alpha] # set opacity of the plotted area

    # Save figure ***Not yet written***
    def saveFig(self):
        pass
    
    # Save parameters in 'params.txt'
    def saveParams(self, b):
        param = [self.a, self.b, self.c, self.d]
        f = open('params.txt', 'a+')
        s = ''
        for p in param:
            s += str(p) + '\t'
        s += '\n'
        f.write(s)
        
        
        if saveBtn.style.button_color == '#ffa07a':
            saveBtn.style.button_color = '#32cd32'
        else:
            saveBtn.style.button_color = '#ffa07a'

c = curve()
```

```python
# Slider GUI 
slider = interactive(c.pattern,
         a=widgets.IntSlider(min=-20, max=20, value=6),
         b=widgets.IntSlider(min=-20, max=20, value=14),
         c=widgets.FloatSlider(min=-10, max=10, step=0.1, value=2),
         d=widgets.FloatSlider(min=-10, max=10, step=0.1, value=3),
         alpha=widgets.FloatSlider(min=0, max=1, step=0.01, value=1.0),
         color=['red', 'green', 'blue', 'purple', '#ff3333', '#a369af', '#073642', '#983101', '#007aff', '#ff9500']
        )

# Button to save the parameters
saveBtn = widgets.Button(description="Save Parameters")
saveBtn.style.button_color = '#ffa07a'
saveBtn.on_click(c.saveParams)

# Display all the GUI widgets
display(HBox([c.fig, VBox([slider, saveBtn])], layout=Layout(align_items = 'center')))
```
